## ライブラリのインポート

```python
import numpy as np
# アイリスの計測データ
from sklearn.datasets import load_iris
# データを扱う
import pandas as pd
# グラフ描画
from matplotlib import pyplot as plt
import japanize_matplotlib
```

## テスト用データの読み込み

```python
iris = load_iris()
iris_data=iris.data
iris_target=np.array(iris.target,dtype=np.float)
```

```python

# アイリスの計測データの各項目
## sepal length (cm)		がく片の長さ
## sepal width (cm)			がく片の幅
## petal length (cm)		花弁の長さ
## petal length (cm)		花弁の幅

# アイリスの計測データを表形式で表示
df_x=pd.DataFrame(iris.data ,columns=iris.feature_names)
df_x.head()
```

```python
df_x.columns.values
```

```python
# アイリスの計測データを表形式で表示
df_y=pd.DataFrame(iris.target ,columns=['target'])
```

```python
df = pd.concat([df_x ,df_y], axis=1)
df.head()
```

```python


# 散布図で表示
st_data = np.array(df.loc[df['target']==0,:])         # Setosa
vc_data = np.array(df.loc[df['target']==1,:])         # Versicolor
vn_data = np.array(df.loc[df['target']==2,:])         # Versinica

#plt.hist([st_data[:, 0], vc_data[:, 0] ,vn_data[:, 0]])
plt.hist([st_data[:, 0]])
#plt.xlabel("target")
#plt.ylabel("sepal length (cm)") # ガクの長さ



plt.show()

```

```python
# 散布図で表示
st_data = df.loc[df['target']==0,:]         # Setosa
vc_data = df.loc[df['target']==1,:]         # Versicolor
vn_data = df.loc[df['target']==2,:]         # Versinica

i = 0
plt.figure(figsize=(20,20))
for col in df_x.columns.values:
    plt.subplot(4, 3, i + 1)  # 2行5列、i+1の位置  (画像表示の指示)
    plt.hist([st_data.loc[:, col]])
    if (i + 1) == 1:
        plt.title('Setosa')
    plt.ylabel(col)
    i += 1
    plt.subplot(4, 3, i + 1)  # 2行5列、i+1の位置  (画像表示の指示)
    plt.hist([vc_data.loc[:, col]])
    if (i + 1) == 2:
        plt.title('Versicolor')
    i += 1
    plt.subplot(4, 3, i + 1)  # 2行5列、i+1の位置  (画像表示の指示)
    plt.hist([vn_data.loc[:, col]])
    if (i + 1) == 3:
        plt.title('Versinica')
    i += 1

plt.show()
```
