## ライブラリのインポート

```python
import numpy as np
# サイキットラーン　（メモリの節約のため、sklearnの中からLinearRegressionのみをインポートする
import sklearn
from sklearn.linear_model import LinearRegression
# ボストンの住宅情報
from sklearn.datasets import load_boston
# 訓練データと検証データの分割
from sklearn.model_selection import train_test_split
# データを扱う
import pandas as pd
# グラフ描画
from matplotlib import pyplot as plt
import japanize_matplotlib
```

## テスト用データの読み込み

```python
boston = load_boston()
# ボストンの家賃のデータの各項目
## CRIM		町ごとの一人当たりの犯罪率
## ZN		宅地の比率が25,000平方フィートを超える敷地に区画されている。
## INDUS	町当たりの非小売業エーカーの割合
## CHAS		チャーリーズ川ダミー変数（川の境界にある場合は1、それ以外の場合は0）
## NOX		一酸化窒素濃度（1000万分の1）
## RM		1住戸あたりの平均部屋数
## AGE		1940年以前に建設された所有占有ユニットの年齢比率
## DIS		5つのボストンの雇用センターまでの加重距離
## RAD		ラジアルハイウェイへのアクセス可能性の指標
## TAX		10,000ドルあたりの税全額固定資産税率
## PTRATIO	生徒教師の比率
## B		町における黒人の割合
## LSTAT	人口当たり地位が低い率
## MEDV		1000ドルでの所有者居住住宅の中央値

# ボストンの家賃のデータを表形式で表示
df=pd.DataFrame(boston.data ,columns=boston.feature_names)
df.head()
```

## 相関係数の算出、描画

```python
# 相関係数の算出 (０は相関がない、１・－１のどちらかに近づけば相関がある)
df_chk=df
df_chk['TARGET']=boston.target
df_chk.corr()
```

```python
## matplotlib が使えることが前提
%matplotlib inline
import seaborn as sns

#ペアプロット図
# 相関関係を目視で確認
sns.pairplot(df_chk)
```

```python
# 分布の確認　（正規分布となっていることを確認）
sns.distplot(df['RM'])
```

## テストデータの中から単回帰分析に使用する項目を定義する

```python
# 入力データ
x = df.iloc[:,5:6]
x.head()
```

```python
# 結果データ
y = pd.DataFrame(boston.target ,columns=['TARGET'])
y.head()
```

```python
# 分布の確認　（正規分布となっていることを確認）
sns.distplot(y['TARGET'])
```

```python
# 散布図（入力データと出力データを描画する）
# この図に
plt.scatter(x ,y ,s=1,c='blue')
plt.show
```

## 訓練データと検証データに分割する

訓練データを使用して学習する。 \
学習した結果を検証データを使用して精度が高いか確認する \
（訓練したデータで一番精度が高くなるように学習するため、訓練データでは \
　それ以上精度を上げることができない。 \
　さらに、将来発生するデータで予測に使いたいので、検証データでの精度で \
　問題がないか判断する）

```python
# 訓練データと検証データの分割
X_train,X_test,y_train,y_test=train_test_split(x, y, test_size=0.4, random_state=1) #random_stateは再現性を確保するために指定する
```

## 学習する

```python
#線回帰モデルの定義
model = LinearRegression()
# モデルの学習
model.fit(X_train ,y_train)
```

```python
# 求まったモデルの結果
a = model.coef_               ##傾き
b = model.intercept_          ##切片
print('a',a)
print('b',b)
```

```python
# 散布図 （予測結果を描画する）
y_hat= a * x + b
plt.scatter(x,y,s=1,c='blue',label= 'data')
plt.plot(x , y_hat, c='red' ,label='予測値')


plt.legend(loc='lower right')
plt.title('ボストンの住宅価格')    
plt.xlabel('RM')   
plt.ylabel('TARGET')   

plt.show
```

```python
# 散布図 （予測結果を描画する）
y_hat_train= a * X_train + b
plt.scatter(X_train,y_train,s=1,c='blue',label= 'data')
plt.plot(X_train , y_hat_train, c='red' ,label='予測値')

plt.legend(loc='lower right')
plt.title('ボストンの住宅価格(訓練データのみ)')    
plt.xlabel('RM')   
plt.ylabel('TARGET')   

plt.show
```

```python
# 検証 <-訓練データ 
model.score(X_train ,y_train)
```

## 検証データを使って予測値の検証

```python
# 検証 <-検証データ(検証データと値が近いなら信用できる。どこまでの差異を近いとするかはそれぞれ判断する必要がある。)
model.score(X_test,y_test)
```

# 外れ値除去
（イレギュラーな値を見つけて、モデルを求める際の入力データから除く）

正規分布に従っていることが、３σ法で除去できる条件である。

```python
# x と y の値をひとつの変数に格納する
df2 = df
#個数カラムを落とす (列の削除)
df2 = df2.drop('CRIM', axis='columns')
df2 = df2.drop('ZN', axis='columns')
df2 = df2.drop('INDUS', axis='columns')
df2 = df2.drop('CHAS', axis='columns')
df2 = df2.drop('NOX', axis='columns')
df2 = df2.drop('AGE', axis='columns')
df2 = df2.drop('DIS', axis='columns')
df2 = df2.drop('RAD', axis='columns')
df2 = df2.drop('TAX', axis='columns')
df2 = df2.drop('PTRATIO', axis='columns')
df2 = df2.drop('B', axis='columns')
df2 = df2.drop('LSTAT', axis='columns')
df2['TARGET'] = y
df_bk=df2
df2.head()
```

```python
# 標準偏差(Standard deviation)
sigma=df2.std()
sigma
```

```python
# 平均
mean = df2.mean()
mean
```

### 入力データを基準に外れ値を除去する

```python
# 外れ値の下限 (平均の値から、標準偏差の２倍の差がある）
low=mean['RM']-2.5* sigma['RM']
low
```

```python
# 外れ値の上限 (平均の値から、標準偏差の２倍差異がある）
high=mean['RM']+2.5 * sigma['RM']
high
```

```python
# ３σ法によって、外れ値を除去する　（全体の3%にあたる情報を削除する）
df2=df2[(df2['RM']>low) & (df2['RM']<high)]
df2.head()
```

```python
print('元のデータセット',len(df_bk))
```

```python
print('外れ値を除去後のデータセット',len(df2))
```

### 出力データを基準に外れ値を除去する

```python
# 外れ値の下限 (平均の値から、標準偏差の２倍の差がある）
low=mean['TARGET']-2.3* sigma['TARGET']
low
```

```python
# 外れ値の上限 (平均の値から、標準偏差の２倍差異がある）
high=mean['TARGET']+2.3 * sigma['TARGET']
high
```

```python
# ３σ法によって、外れ値を除去する　（全体の3%にあたる情報を削除する）
df2=df2[(df2['TARGET']>low) & (df2['TARGET']<high)]
df2.head()
```

```python
print('元のデータセット',len(df_bk))
```

```python
print('外れ値を除去後のデータセット',len(df2))
```

```python
x2 = df2.iloc[:,0:1]
x2.head()
```

```python
y2 = df2.iloc[:,-1:]
y2.head()
```

```python
# 訓練データと検証データの分割
#random_stateは再現性を確保するために指定する
X2_train,X2_test,y2_train,y2_test=train_test_split(x2, y2, test_size=0.4, random_state=1) 
```

## 分析

```python
#線回帰モデルの定義
model2 = LinearRegression()
# モデルの学習
model2.fit(X2_train ,y2_train)
```

```python
# 求まったモデルの結果
a2 = model2.coef_               ##傾き
b2 = model2.intercept_          ##切片
print('a',a2)
print('b',b2)
```

```python
# 散布図 （予測結果を描画する）
y_hat2 = a2 * x + b2
plt.scatter(x,y,s=1,c='blue',label= 'data')
plt.plot(x , y_hat, c='black' ,label='予測値')
plt.plot(x , y_hat2, c='red' ,label='予測値（外れ値除去）')

plt.legend(loc='lower right')
plt.title('ボストンの住宅価格') 
plt.xlabel('RM')   
plt.ylabel('TARGET')   

plt.show
```

```python
# 散布図 （予測結果を描画する）
y_hat2 = a2 * x2 + b2
plt.scatter(x2,y2,s=1,c='blue',label= 'data')
plt.plot(x2 , y_hat2, c='red' ,label='予測値（外れ値除去）')

plt.legend(loc='lower right')
plt.title('ボストンの住宅価格（外れ値除去後）') 
plt.xlabel('RM')   
plt.ylabel('TARGET')   

plt.show
```

```python
# 散布図 （予測結果を描画する）
y_hat2_train = a2 * X2_train + b2
plt.scatter(X2_train ,y2_train ,s=1,c='blue',label= 'data')
plt.plot(X2_train , y_hat2_train, c='red' ,label='予測値（外れ値除去）')

plt.legend(loc='lower right')
plt.title('ボストンの住宅価格（外れ値除去後訓練データ）') 
plt.xlabel('RM')   
plt.ylabel('TARGET')   

plt.show
```

```python
# 散布図 （予測結果を描画する）
y_hat2_test = a2 * X2_test + b2
plt.scatter(X2_test ,y2_test ,s=1,c='blue',label= 'data')
plt.plot(X2_test , y_hat2_test, c='red' ,label='予測値（外れ値除去）')

plt.legend(loc='lower right')
plt.title('ボストンの住宅価格（外れ値除去後検証データ）') 
plt.xlabel('RM')   
plt.ylabel('TARGET')   

plt.show
```

```python
# 検証 <-訓練データ 
model2.score(X2_train ,y2_train)
```
