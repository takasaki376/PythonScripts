```python
def Hex2rgb(hex):
    return int(hex ,16)
```

```python
r = np.arange(0 , 255 , 2)
#r.append(255)
print(r)
```

```python
sample2 = []
for idx in range(len(sample)):
    int1 = Hex2rgb(sample[idx][0])
    int2 = Hex2rgb(sample[idx][1])
    int3 = Hex2rgb(sample[idx][2])
    season = sample[idx][3]
    sample2.append([int1 ,int2, int3 ,season])
print(sample2)
```

```python
result = []
count = 0
count2= 0
for rIdx in range(len(r)):
    for bIdx in range(len(r)):
        for gIdx in range(len(r)):
            for idx in range(len(sample2)):
                count += 1
                x = sample2[idx][0:3]
                y = [r[rIdx] , r[bIdx] , r[gIdx]]
                
                rmse = np.sqrt(mean_squared_error(x, y))
                if (rmse <= 35):
                    count2 += 1
                    result.append([count ,sample2[idx][0],sample2[idx][1],sample2[idx][2],sample2[idx][3] , r[rIdx] , r[bIdx] , r[gIdx], rmse ])
            
print(count)
print(count2)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(result , columns=['No', 'sample_r', 'sample_b', 'sample_g', 'season','color_r', 'color_b', 'color_g', 'rmse'])
```
