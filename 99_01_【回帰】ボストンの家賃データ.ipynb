## ライブラリのインポート

```python
import numpy as np
# ボストンの住宅情報
from sklearn.datasets import load_boston
# データを扱う
import pandas as pd
# グラフ描画
from matplotlib import pyplot as plt
import japanize_matplotlib
```

## テスト用データの読み込み

```python
boston = load_boston()
```

```python
# 説明
print(boston.DESCR)
```

```python
boston
```

```python
#入力データを表形式で表示する

# ボストンの家賃のデータの各項目
## 1.  CRIM      町ごとの一人当たりの犯罪率
## 2.  ZN        宅地の比率が25,000平方フィートを超える敷地に区画されている。
## 3.  INDUS     町当たりの非小売業エーカーの割合
## 4.  CHAS      チャーリーズ川ダミー変数（川の境界にある場合は1、それ以外の場合は0）
## 5.  NOX       一酸化窒素濃度（1000万分の1）
## 6.  RM        1住戸あたりの平均部屋数
## 7.  AGE       1940年以前に建設された所有占有ユニットの年齢比率
## 8.  DIS       5つのボストンの雇用センターまでの加重距離
## 9.  RAD       ラジアルハイウェイへのアクセス可能性の指標
## 10.  TAX      10,000ドルあたりの税全額固定資産税率
## 11.  PTRATIO  生徒教師の比率
## 12.  B        町における黒人の割合
## 13.  LSTAT    人口当たり地位が低い率
## 14.  MEDV     1000ドルでの所有者居住住宅の中央値

# ボストンの家賃のデータを表形式で表示
df_x=pd.DataFrame(boston.data ,columns=boston.feature_names)
df_x.head(50)
```

```python
# ボストンの家賃のデータを表形式で表示
df_y=pd.DataFrame(boston.target ,columns=['yatin'])
df_y.head()
```

```python
# 相関係数の算出 (０は相関がない、１・－１のどちらかに近づけば相関がある)
df_x.corr()
```
