```python
import numpy as np
# ボストンの住宅情報
from sklearn.datasets import load_boston
# データを扱う
import pandas as pd
```

## テスト用データの読み込み

```python
boston = load_boston()
#入力データを表形式で表示する

# ボストンの家賃のデータの各項目
## 1.  CRIM      町ごとの一人当たりの犯罪率
## 2.  ZN        宅地の比率が25,000平方フィートを超える敷地に区画されている。
## 3.  INDUS     町当たりの非小売業エーカーの割合
## 4.  CHAS      チャーリーズ川ダミー変数（川の境界にある場合は1、それ以外の場合は0）
## 5.  NOX       一酸化窒素濃度（1000万分の1）
## 6.  RM        1住戸あたりの平均部屋数
## 7.  AGE       1940年以前に建設された所有占有ユニットの年齢比率
## 8.  DIS       5つのボストンの雇用センターまでの加重距離
## 9.  RAD       ラジアルハイウェイへのアクセス可能性の指標
## 10.  TAX      10,000ドルあたりの税全額固定資産税率
## 11.  PTRATIO  生徒教師の比率
## 12.  B        町における黒人の割合
## 13.  LSTAT    人口当たり地位が低い率
## 14.  MEDV     1000ドルでの所有者居住住宅の中央値

# ボストンの家賃のデータを表形式で表示
df=pd.DataFrame(boston.data ,columns=boston.feature_names)
df.head()
```

# データのマージ

### キー項目で二つのデータをマージする

```python
data =[
    [1   ,'01',   5],
    [2   ,'01',   8],
    [3   ,'02',   1],
    [4   ,'02',   3],
    [5   ,'03',   4],
    [6   ,'03',   7]
]
```

```python
mastar =[
    ['01', 'apple'  ,  100],
    ['02', 'orrange',  155],
    ['04', 'peach',  155]
]
```

```python
df_data = pd.DataFrame(data, columns=['id', 'cd', 'qty'])
df_data.head()
```

```python
df_mastar = pd.DataFrame(mastar, columns=['cd','name', 'price'])
df_mastar.head()
```

```python
#  結合のキーを"cd"にして、内部結合
df_merge = pd.merge(df_data , df_mastar, on='cd' , how='inner')
df_merge.head(10)
```

```python
#  結合のキーを"cd"にして、df_dataに存在するレコードを出力する
df_merge = pd.merge(df_data , df_mastar, on='cd' , how='left')
df_merge.head(10)
```

```python
#  結合のキーを"cd"にして、df_mastarに存在するレコードを出力する
df_merge = pd.merge(df_data , df_mastar, on='cd' , how='right')
df_merge.head(10)
```

```python
#  結合のキーを"cd"にして、どちらか片方にデータが存在していれば出力する
df_merge = pd.merge(df_data , df_mastar, on='cd' , how='outer')
df_merge.head(10)
```
